```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(-3,3,100)
P = np.linspace(0,20,100)
delta = .0001
print ("This program will plot the 'slope' of the quadratic from lecture 5 at various points")

numberoftangents=int(input("How many points would you like to see? "))

points = []
i = 0
while i < numberoftangents:
    try:
        pointi = float(input("Input a number between 0 and 20 : "))
        points.append(pointi)
        i +=1
    except Exception:
        print('Should be a number.  Please try again')


def F(P):
    return 100*P-5*P**2

def f(P):
    return (F(P+delta)-F(P))/delta

def approx(point):
    return F(point)+f(point)*x

plt.plot(P,F(P), label= 'f(x)')

for point in points:
    labelpoint = str(point)
    plt.plot(x+point,approx(point),label= "Tangent when x= " +labelpoint)
    lastlabel1 = str(point)
    lastlabel2 = str(f(point))
    print ("The slope of the tangent line at  " +lastlabel1 +" is : " +lastlabel2)
    print (f(point))
plt.legend()
plt.show()
```
